# Reading

```python
# Install Packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Read in the Datasets

raw = pd.read_csv('College Promise - Alamo Colleges - 03.10.2025.csv') #79 MB
new = pd.read_csv('Alamo Colleges Promise Sample File 23SEP2025.xlsx - Alamo Colleges Promise Sample F.csv') #15 MB
```

# Cleaning (Duplicates)

```python
# Show duplicates in the raw dataset

print(f"Number of duplicate Student_IDs in raw: {raw.duplicated(subset=['Student_ID']).sum()}")
print("\nRows with duplicate Student_IDs in raw:")
raw[raw.duplicated(subset=['Student_ID'], keep=False)].sort_values('Student_ID') # keep=False sets duplicated values as "true"
```

```python
# Show duplicates in the new dataset

print(f"Number of duplicate Student_IDs in new: {new.duplicated(subset=['Student_ID']).sum()}")
print("\nRows with duplicate Student_IDs in new:")
new[new.duplicated(subset=['Student_ID'], keep=False)].sort_values('Student_ID') # keep=False sets duplicated values as "true"
```

```python
# Tag duplicates (the second occurence of each unique Student_ID) in both datasets with "_dup"

# The only issue with this is we don't know if the non-"_dup" and "_dup" Student-IDs match in the raw and new datasets.

raw_tagged = raw.copy()
raw_tagged['Student_ID'] = raw_tagged['Student_ID'].astype(str)
raw_tagged['counter'] = raw_tagged.groupby('Student_ID').cumcount() + 1
print(raw_tagged['counter'].unique()) # prints the unique values of Student_ID counts (1 = unique, 2 = one duplicate)

new_tagged = new.copy()
new_tagged['Student_ID'] = new_tagged['Student_ID'].astype(str)
new_tagged['counter'] = new_tagged.groupby('Student_ID').cumcount() + 1
print(new_tagged['counter'].unique()) # prints the unique values of Student_ID counts (1 = unique, 2 = one duplicate)

raw_tagged['Student_ID'] = np.where( # np.where(condition, value_if_true, value_if_false)
    raw_tagged['counter'] == 2,
    raw_tagged['Student_ID'] + '_dup',
    raw_tagged['Student_ID']
)

new_tagged['Student_ID'] = np.where(
    new_tagged['counter'] == 2,
    new_tagged['Student_ID'] + '_dup',
    new_tagged['Student_ID']
)
```

```python
# Check for Duplicate Student_IDs (there should be 0)

print(raw_tagged[raw_tagged.duplicated()])

print(new_tagged[new_tagged.duplicated()])
```

# Joining

```python
# Join on Student_ID

# need to narrow down to specific variable(s) in the new_tagged dataset

master = pd.merge(raw_tagged, new_tagged, on='Student_ID', how = 'left')
```

```python
# Check that master has the same number of rows as raw and new

len(raw) == len(new) == len(master)
```

```python
# For variables shared across the RAW and NEW datasets,
# the following code will select the dataset which has less missingness for each corresponding variable.
# We can use the selections as a guide in building the final dataset: df.

# Academic Cohort
if (raw['FTIC_Cohort'].isna().sum())/len(raw['FTIC_Cohort']) < (new['FTIC_Cohort'].isna().sum())/len(new['FTIC_Cohort']):
    print('Raw Academic Cohort')
else:
    print('New Academic Cohort')

# Promise Program Status
if (raw['PS_Treatment_Control_1'].isna().sum())/len(raw['PS_Treatment_Control_1']) < (new['FTIC_Status'].isna().sum())/len(new['FTIC_Status']):
    print('Raw Promise Program Status')
else:
    print('New Promise Program Status')

print("Raw Fall-to-Fall Registration [No New]")

# Gender
if (raw['Gender_1'].isna().sum())/len(raw['Gender_1']) < (new['Gender'].isna().sum())/len(new['Gender']):
    print('Raw Gender')
else:
    print('New Gender')

# Academic Disadvantage
if (raw['Academically_Disadvantaged_1'].isna().sum())/len(raw['Academically_Disadvantaged_1']) < (new['Acad_Disadv'].isna().sum())/len(new['Acad_Disadv']):
    print('Raw Academic Disadvantage')
else:
    print('New Academic Disadvantage')

print("Raw Enrollment Status [No New]")

# TSIA Math Readiness
# The NEW dataset includes TSIA 1 and TSIA 2, so the proportions were summed in order to do the comparison.
if (raw['TSIA_Math_College_Readiness_1'].isna().sum())/len(raw['TSIA_Math_College_Readiness_1']) < ((new['TSI_MATH_SCORE'].isna().sum())/len(new['TSI_MATH_SCORE']) + new['TSI_MATH_2_SCORE'].isna().sum())/len(new['TSI_MATH_2_SCORE']):
    print('Raw TSIA')
else:
    print('New TSIA')
print("Though the RAW TSIA data has less missingness, we will also include the NEW TSIA data as it accounts for TSIA 1 and TSIA 2.")

print("Raw Age [No New]")

print("Raw Productive Credit Ratio [No New]")

# Hours Earned (2020)
if (raw['Fall_2020_HOURS_EARNED'].isna().sum())/len(raw['Fall_2020_HOURS_EARNED']) < (new['Fall_2020_HOURS_EARNED'].isna().sum())/len(new['Fall_2020_HOURS_EARNED']):
    print('Raw Hours Earned (2020)')
else:
    print('New Hours Earned (2020)')

# Hours Earned (2021)
if (raw['Fall_2021_HOURS_EARNED'].isna().sum())/len(raw['Fall_2021_HOURS_EARNED']) < (new['Fall_2021_HOURS_EARNED'].isna().sum())/len(new['Fall_2021_HOURS_EARNED']):
    print('Raw Hours Earned (2021)')
else:
    print('New Hours Earned (2021)')

# Hours Earned (2022)
if (raw['Fall_2022_HOURS_EARNED'].isna().sum())/len(raw['Fall_2022_HOURS_EARNED']) < (new['Fall_2022_HOURS_EARNED'].isna().sum())/len(new['Fall_2022_HOURS_EARNED']):
    print('Raw Hours Earned (2022)')
else:
    print('New Hours Earned (2022)')

# Hours Earned (2023)
if (raw['Fall_2023_HOURS_EARNED'].isna().sum())/len(raw['Fall_2023_HOURS_EARNED']) < (new['Fall_2023_HOURS_EARNED'].isna().sum())/len(new['Fall_2023_HOURS_EARNED']):
    print('Raw Hours Earned (2023)')
else:
    print('New Hours Earned (2023)')

# Hours Earned (2024)
if (raw['Fall_2024_HOURS_EARNED'].isna().sum())/len(raw['Fall_2024_HOURS_EARNED']) < (new['Fall_2024_HOURS_EARNED'].isna().sum())/len(new['Fall_2024_HOURS_EARNED']):
    print('Raw Hours Earned (2024)')
else:
    print('New Hours Earned (2024)')

print("Raw Academic Eligibility [No New]")
```

# Recoding Variables

```python
# Filtering Variable: Academic Cohort
print(master['FTIC_Cohort_y'].unique())

# 2020 = 1, 2021 = 2, 2022 = 3, 2023 = 4, 2024 = 5
cohort_map = {'Fall 2020 FTIC Cohort': 1,
              'Fall 2021 FTIC Cohort': 2,
              'Fall 2022 FTIC Cohort': 3,
              'Fall 2023 FTIC Cohort': 4,
              'Fall 2024 FTIC Cohort': 5}
master['Cohort_Factor'] = master['FTIC_Cohort_y'].map(cohort_map)
```

```python
# Treatment: Promise Program Status
print(master['FTIC_Status_y'].unique())

# Promise FTIC = 1, non-Promise FTIC = 0
promise_map = {'Promise FTIC': 1,
               'non-Promise FTIC': 0}
master['Treatment'] = master['FTIC_Status_y'].map(promise_map)
```

```python
# Dependent Variable: Fall-to-Fall Retention

print(master['Fall_2020_F_F_Registered_2'].unique())
print(master['Fall_2021_F_F_Registered_2'].unique())
print(master['Fall_2022_F_F_Registered_2'].unique())
print(master['Fall_2023_F_F_Registered_2'].unique())

# Promise FTIC = 1, non-Promise FTIC = 0
retention_map = {'Yes': 1,
                  'No': 0}
master['Fall_2020_F_F_Registered_Factor'] = master['Fall_2020_F_F_Registered_2'].map(retention_map)
master['Fall_2021_F_F_Registered_Factor'] = master['Fall_2021_F_F_Registered_2'].map(retention_map)
master['Fall_2022_F_F_Registered_Factor'] = master['Fall_2022_F_F_Registered_2'].map(retention_map)
master['Fall_2023_F_F_Registered_Factor'] = master['Fall_2023_F_F_Registered_2'].map(retention_map)
```

```python
# Gender
print(master['Gender_y'].unique())

# Female = 0, Male = 1
gender_map = {'Female': 0,
              'Male': 1}
master['Gender_Factor'] = master['Gender_y'].map(gender_map)
```

```python
# Academic Disadvantage

print(master['Acad_Disadv_y'].unique())

# Not Academic Disadvanted = 0, Academic Disadvanted = 1
acad_disadv_map = {'Not Academic Disadvantaged': 0,
                   'Academic Disadvantaged': 1}
master['Academically_Disadvantaged_Factor'] = master['Acad_Disadv_y'].map(acad_disadv_map)
```

```python
# Enrollment Status

print(master['Enrollment_Status_1'].unique())

# Part-Time = 0, Full-Time = 1
enrollment_map = {'Part-Time': 0,
                  'Full-Time': 1}
master['Enrollment_Status_Factor'] = master['Enrollment_Status_1'].map(enrollment_map)
```

```python
# TSIA Math Score

# RAW TSIA Data
print(master['TSIA_Math_College_Readiness_1'].unique())
# Not College Ready = 0, College Ready = 1
tsia_map = {'Not College Ready': 0,
            'College Ready': 1}
master['TSIA_Math_College_Readiness_Factor'] = master['TSIA_Math_College_Readiness_1'].map(tsia_map)

# TSI 1: Minimum score of 350
master['TSI_MATH_1_Factor'] = pd.cut(master['TSI_MATH_SCORE_y'],
                                    bins = [0, 349, 999],
                                    labels = ['0', '1'])

# TSI 2: Minimum score of 950
master['TSI_MATH_2_Factor'] = pd.cut(master['TSI_MATH_2_SCORE_y'],
                                    bins = [0, 949, 999],
                                    labels = ['0', '1'])
```

```python
# Age

print(master['Age_At_Entering_Cohort_2'].unique())

# Aged 24 and Under = 0, Aged 25 and Older = 1
age_map = {'Aged 24 and Under': 0,
           'Aged 25 and Older': 1}
master['Age_Factor'] = master['Age_At_Entering_Cohort_2'].map(age_map)
```

```python
# Academic Eligibility

print(master['Fall_2020_Academically_Eligible_1'].unique())
print(master['Fall_2021_Academically_Eligible_1'].unique())
print(master['Fall_2022_Academically_Eligible_1'].unique())
print(master['Fall_2023_Academically_Eligible_1'].unique())
print(master['Fall_2024_Academically_Eligible_1'].unique())

eligibility_map = {'No': 0,
                   'Yes': 1,
                   ' ': None}
master['Fall_2020_Academically_Eligible_Factor'] = master['Fall_2020_Academically_Eligible_1'].map(eligibility_map)
master['Fall_2022_Academically_Eligible_Factor'] = master['Fall_2022_Academically_Eligible_1'].map(eligibility_map)
master['Fall_2023_Academically_Eligible_Factor'] = master['Fall_2023_Academically_Eligible_1'].map(eligibility_map)
```

# Creating the Dataset

```python
df = master[['Student_ID',
             'Cohort_Factor',
             'Treatment',
             'Fall_2020_F_F_Registered_Factor',
             'Fall_2021_F_F_Registered_Factor',
             'Fall_2022_F_F_Registered_Factor',
             'Fall_2023_F_F_Registered_Factor',
             'Gender_Factor',
             'Academically_Disadvantaged_Factor',
             'Enrollment_Status_Factor',
             'TSIA_Math_College_Readiness_Factor',
             'TSI_MATH_1_Factor',
             'TSI_MATH_2_Factor',
             'Age_Factor',
             'Fall_2020_Productive_Credit_Ratio_1',
             'Fall_2021_Productive_Credit_Ratio_1',
             'Fall_2022_Productive_Credit_Ratio_1',
             'Fall_2023_Productive_Credit_Ratio_1',
             'Fall_2020_HOURS_EARNED_x',
             'Fall_2021_HOURS_EARNED_x',
             'Fall_2022_HOURS_EARNED_x',
             'Fall_2023_HOURS_EARNED_x',
             'Fall_2020_Academically_Eligible_Factor',
             'Fall_2021_Academically_Eligible_1',
             'Fall_2022_Academically_Eligible_Factor',
             'Fall_2023_Academically_Eligible_Factor',
             'Fall_2024_Academically_Eligible_1']]
```

```python
# Variance Inflation Factor

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_data
```
